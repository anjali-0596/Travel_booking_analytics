```python
import pandas as pd
import numpy as np
df = pd.read_csv("hotel_bookings.csv")
df.head()
df.shape
df.columns
df.info()
df['children'] = df['children'].fillna(0)
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
df.isnull().sum()
df['country'] = df['country'].fillna('Unknown')
df.isnull().sum()
cancellation_rate = df['is_canceled'].mean() * 100
print(f"Overall Cancellation Rate: {cancellation_rate: .2f}%")
hotel_cancellation = df.groupby('hotel')['is_canceled'].mean() * 100
print(hotel_cancellation)
df.groupby('is_canceled')['lead_time'].mean()
df.groupby('customer_type')['is_canceled'].mean() * 100
df.groupby('market_segment')['is_canceled'].mean() * 100
df.groupby('hotel')['adr'].mean()
df['arrival_date_month'].value_counts()
df.groupby('reserved_room_type')['adr'].mean().sort_values(ascending=False)
```
